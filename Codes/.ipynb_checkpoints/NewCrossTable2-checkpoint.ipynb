```python
#Extract All of Part1

rank_number = len(study_json["StudyFieldsResponse"]["StudyFields"])
fields_number =len( study_json["StudyFieldsResponse"]["StudyFields"][0])
print (rank_number)

#Extracting from multiple files

for trial_batch in range (0,2):

    trial_batch = 0
    trial_path="json2/field_usa"+str(trial_batch)+"_part_1.json"

    with open(trial_path) as jsonfile:
        study_json = json.load(jsonfile)

    rank_number = len(study_json["StudyFieldsResponse"]["StudyFields"])
    fields_number =len( study_json["StudyFieldsResponse"]["StudyFields"][0])

    for x in range (0,rank_number):

    #     if AgreementOtherDetails.append(study_json['StudyFieldsResponse']['StudyFields'][x]['AgreementOtherDetails']='':
    #         AgreementOtherDetails.append(0)
    #     else:
    #         AgreementOtherDetails.append(study_json['StudyFieldsResponse']['StudyFields'][x]['AgreementOtherDetails'][0]

        study_content= study_json['StudyFieldsResponse']['StudyFields'][x]

        if len(study_content['NCTId']) ==0 :
            NCTId.append(0)
        if len(study_content['NCTId']) ==1 :
            NCTId.append(study_content['NCTId'][0]) 
        elif len(study_content['NCTId']) > 1:
            NCTId.append(study_content['NCTId'])  
            
        if len(study_content['ArmGroupInterventionName']) ==0 :
            ArmGroupInterventionName.append(0)
        if len(study_content['ArmGroupInterventionName']) ==1 :
            ArmGroupInterventionName.append(study_content['ArmGroupInterventionName'][0]) 
        elif len(study_content['ArmGroupInterventionName']) > 1:
            ArmGroupInterventionName.append(study_content['ArmGroupInterventionName'])  

        if len(study_content['ArmGroupType']) ==0 :
            ArmGroupType.append(0)
        if len(study_content['ArmGroupType']) ==1 :
            ArmGroupType.append(study_content['ArmGroupType'][0]) 
        elif len(study_content['ArmGroupType']) > 1:
            ArmGroupType.append(study_content['ArmGroupType'])  
```

```python
#Create Dataframe from list
dict1 = { part1[0]:NCTId,

part1[1]:ArmGroupInterventionName,ArmGroupInterventionName

part1[2]:ArmGroupType}

df1 = pd.DataFrame(dict1)

#df1.to_csv('fields_all_part1.csv')
```

```python
df1
```

```python
def explodeTable(df_name, column_name):
    tb = df_name[['NCTId', column_name]].copy()
    explode_tb = tb.explode(column_name)
    explode_tb = explode_tb.replace(0, "-")
    explode_tb = explode_tb.sort_values(by=[column_name], ascending=False)
    fr = len(explode_tb) - explode_tb[column_name].str.contains('-', na=False).sum()
    print('feature_row', fr)
    feature_tb = explode_tb.head(fr)
    #.rename(columns={column_name:'features'})

    return feature_tb
```

```python
tb1= explodeTable(df1, 'ArmGroupInterventionName')
tb1
```

```python
tb1['feature'] = tb1['ArmGroupInterventionName'].str.split(' ').str[0]
tb1 = tb1.drop(['ArmGroupInterventionName'], axis=1)
tb1 = tb1.rename(columns={'feature':'ArmGroupInterventionName'})
tb1
```

```python
# Creating cross table
Cross_Tb1 = pd.crosstab(index=tb1['NCTId'], columns = tb1['ArmGroupInterventionName'])
Cross_Tb1
```

```python
Cross_Tb1.to_csv('Cross_Tb1.csv')
```

```python
#Creating 0 and 1 matrix table
Cross_Tb1[Cross_Tb1>1]=1
Cross_Tb1.reset_index()
Cross_Tb1
```
