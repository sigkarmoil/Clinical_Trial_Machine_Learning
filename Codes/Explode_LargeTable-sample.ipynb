```python
# Dependencies
import json
import os
import pandas as pd
```

```python
#open json
object_path = 'sample_data.json'
with open(object_path) as jsonfile:
       data_json = json.load(jsonfile)
df_from_json = pd.DataFrame(data_json)
```

```python
def explodeTable(df_name, column_name):
    tb = df_name[['NCTId', column_name]].copy()
    tb = tb.explode(column_name)
    #create cross table
    Cross_Tb = pd.crosstab(index=tb['NCTId'], columns = tb[column_name])
    #Creating 0 and 1 matrix table
    Cross_Tb[Cross_Tb>1]=1
    Cross_Tb.reset_index()
    
    return Cross_Tb
```

```python
FullTable_df = df_from_json.copy()
FullTable_df.drop(FullTable_df.iloc[:, 16:20], axis=1, inplace=True)
FullTable_df.drop(FullTable_df.iloc[:, 1:10], axis=1, inplace=True)
FullTable_df.drop(['InterventionType','OverallStatus','Phase','StdAge'], axis = 1, inplace=True)
FullTable_df.sort_values('NCTId', inplace=True)
FullTable_df.reset_index(drop=True, inplace=True)
FullTable_df = FullTable_df.rename(columns={'NCTId':'ID'})
FullTable_df 
```

```python
Cross_Tb2= explodeTable(df_from_json, 'ArmGroupType')
# Cross_Tb2.to_csv('CSVTables/Cross_Tb2.csv')
Cross_Tb2.reset_index(inplace=True)
Cross_Tb2.drop(Cross_Tb2.columns[[1]], axis=1, inplace=True)
Cross_Tb2
```

```python
FullTable_df = FullTable_df.merge(Cross_Tb2, left_on='ID', right_on='NCTId')
FullTable_df.drop(['NCTId'], axis = 1, inplace=True)
FullTable_df
```
