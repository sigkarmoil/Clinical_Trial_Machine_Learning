```python
# Dependencies
import json
import os
import pandas as pd
```

```python
#open json
object_path = 'renamed_all_data.json'
with open(object_path) as jsonfile:
       data_json = json.load(jsonfile)
df_from_json = pd.DataFrame(data_json)
```

```python
df_from_json
```

```python
tb1 = df_from_json[['NCTId', 'ArmGroupInterventionName']].copy()
tb1 = tb1.explode('ArmGroupInterventionName')
tb1['feature'] = tb1['ArmGroupInterventionName'].str.split(':').str[0]
tb1 = tb1.drop(['ArmGroupInterventionName'], axis=1)
tb1 = tb1.rename(columns={'feature':'ArmGroupInterventionName'})
tb1 = tb1.sort_values(by=['ArmGroupInterventionName'], ascending=False)
tb1
```

```python
Cross_Tb1 = pd.crosstab(index=tb1['NCTId'], columns = tb1['ArmGroupInterventionName'])
#Creating 0 and 1 matrix table
Cross_Tb1[Cross_Tb1>1]=1
Cross_Tb1.reset_index()
Cross_Tb1
```

```python
Cross_Tb1.to_csv('CSVTables/Cross_Tb1.csv')
```
