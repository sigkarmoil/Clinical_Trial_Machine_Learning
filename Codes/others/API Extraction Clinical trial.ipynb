```python
import requests
import pprint
import json
import time
```

```python
# min_rnk = 1
# max_rnk = 100
```

```python
for a in range (310,1390):
    min_rnk = 1 + 100*a
    max_rnk = 100 + 100*a
    print (min_rnk)
    print (max_rnk)
    time.sleep(5)
    
    url_extract = f"https://clinicaltrials.gov/api/query/full_studies?expr=United+States&min_rnk={min_rnk}&max_rnk={max_rnk}&fmt=json" 
    response = requests.get(url_extract)
    response_json = response.json()
    with open(f'usa{a}.json', 'w') as json_file:
        json.dump(response_json, json_file)
```
